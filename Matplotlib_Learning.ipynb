# Subplots

```python
# A diff way to plot graphs
batter.head()
```

```python
plt.scatter(batter['avg'],batter['strike_rate'])
plt.title('SomeThing')
plt.xlabel("Avg")
plt.ylabel("Strike_rate")
```

```python
# Other way  ------> (using this approach we can create multiple graph side by side)
fig,ax = plt.subplots(figsize = (15,6))
ax.scatter(batter['avg'],batter['strike_rate'],marker ='.')
ax.set_title('Something')
ax.set_xlabel('Avg')
ax.set_ylabel('Strik_rate')
```

```python
fig,ax = plt.subplots(nrows=2,ncols=1,sharex=True,figsize=(10,10))
ax[0].scatter(batter['avg'],batter['strike_rate'],marker ='.')
ax[0].set_xlabel('Avg')
ax[0].set_ylabel('Strik_rate')
ax[0].set_title('Avg vs Strik_rate')

ax[1].scatter(batter['avg'],batter['runs'],marker ='.')
ax[1].set_xlabel('Avg')
ax[1].set_ylabel('Runs')
ax[1].set_title('Avg vs Runs')
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(1,2,1)
ax1.scatter(batter['avg'],batter['strike_rate'],marker ='.')
ax2 = fig.add_subplot(1,2,2)
ax1.scatter(batter['avg'],batter['runs'],marker ='.')
```
