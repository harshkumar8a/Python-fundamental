# Series from lists

*   a one-dimensional labeled array capable of holding any data type



```python
# String
country = ["India","USA","China","Nepal","Pakistan"]
pd.Series(country,index=[1,2,3,4,5],name="Country")
```

```python
# integers
number = [1,2,3,4,5]
pd.Series(number,name="Number")
```

```python
# custom index
marks = [67,59,48,35,74]
subject = ["Math","English","Hindi","Science","SST"]
pd.Series(marks,index=subject)
```

# Group By

```python
# Always applied on catogoical Data
```

```python
# In these type of question first we groupby which we want a group , after that
# we decide that on which columns we have to separte then we o aggregate operation
```

```python
movies = pd.read_csv("/content/Top_10000_Movies_IMDb.csv")
deliveries = pd.read_csv("/content/deliveries.csv")
```

```python
deliveries.info()
```

```python
movies.head(1)
```

```python
genres = movies.groupby('Genre')
genres
```

```python
# Applying builtin aggregration functions on group by objects
genres.sum()
```

```python
# find the top 3 genre by total earning
movies.groupby('Genre').sum()['Gross'].sort_values(ascending = False).head(3)
```

```python
movies.groupby('Genre')['Gross'].sum().sort_values(ascending = False).head(3)
```

```python
#fin the genre with highest avg IMDB rating
movies.groupby('Genre')['Rating'].mean().sort_values(ascending = False).head(3)
```

```python
# find the director with most popularity
movies.groupby('Genre')['Votes'].sum().sort_values(ascending = False).head(3)
```

```python
# find total number of groups -> len
len(movies.groupby('Genre'))
```

```python
# find item in each group -> size
movies.groupby('Genre').size()
```

```python
movies['Genre'].value_counts()
```

```python
# first()/last()  --> nth item
genre = movies.groupby('Genre')
genre.first()
genre.last()
genre.nth(6)
```

```python
# get_group --> vs filtering
genre.get_group('Horror')
```

```python
genre.groups
```

```python
genre.describe()
```

```python
# aggr methods
genre.agg(['min','max','sum'])
```

```python
# loop on groups
for groups,data in genre:
    print(data)
```
